```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
import warnings
warnings.filterwarnings('ignore')
```

```python
Int = pd.read_csv('dataset/Internet_Demographic.csv')
Int
```

<h1> 1. Distribution of Birth Rates </h1>

```python
plt.hist(Int['Birth rate'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Birth Rates')
plt.xlabel('Birth Rate')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<h1> 2. Top 5 Countries with Highest Internet Usage </h1>

```python
top_countries = Inter.nlargest(5, 'Internet users')
print(top_countries[['Country Name', 'Internet users']])
```

<h1> 3. Count of Countries by Income Group </h1>

```python
income_group_counts = Int['Income Group'].value_counts()
income_group_counts.plot(kind='bar', color='purple', edgecolor='black')
plt.title('Count of Countries by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Count')
plt.show()
```

<h1>  4. Internet Usage </h1>

```python
Int.boxplot(column=['Internet users'])
plt.title('Box Plot of Internet Users')
plt.ylabel('Internet Users (%)')
plt.show()
```

<h1>  5. Countries with No Internet Access </h1>

<h1> 6. Average Birth Rate </h1>

```python
avg_br = Int['Birth rate'].mean()
print(f'Average Birth Rate: {avg_br}')
```

<h1> 7. Birth Rate vs Internet Usage </h1>

```python
plt.scatter(Int['Birth rate'], Int['Internet users'], alpha=0.7)
plt.title('Birth Rate vs Internet Usage')
plt.xlabel('Birth Rate')
plt.ylabel('Internet Users (%)')
plt.grid(alpha=0.5)
plt.show()
```

<h1> 8. Countries with Birth Rates Above 40 </h1>

```python
high_br = Int[Int['Birth rate'] > 40]
print(high_br[['Country Name', 'Birth rate']])
```

<h1> 9. Proportion of Income Groups </h1>

```python
income_group_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Income Groups')
plt.ylabel('')
plt.show()
```

<h1> 10. Correlation Between Birth Rate and Internet Usage </h1>

```python
corr = Int['Birth rate'].corr(Int['Internet users'])
print(f'Correlation between Birth Rate and Internet Usage: {corr}')
```
